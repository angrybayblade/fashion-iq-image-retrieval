```python
import glob
import json
import cv2
import re

import matplotlib.pyplot as plt
import numpy as np

from tqdm.notebook import tqdm
from gc import collect
from time import sleep

%matplotlib inline
```

```python
SYMBOL_FILTER = re.compile('\W')
```

```python
image = cv2.cvtColor(
    cv2.resize(
        cv2.imread("./data/images/B00BZ8GPVO .jpg"),
        (128,128)
    ),
    cv2.COLOR_BGR2RGB
)

plt.imshow(image)
```

```python
json_files = glob.glob("./data/captions/*.json")
captions_raw = []

for file in json_files:
    with open(file,"r") as file:
        captions_raw += json.load(file)
```

```python
%%time

image_data = []
captions = []
broken = []

for i,caption in enumerate(captions_raw):
    if not i%1000:
        collect()
    try:
        path = f'./data/images/{caption["candidate"]} .jpg'
        image_data += [
            cv2.cvtColor(
                cv2.resize(
                    cv2.imread(path),
                    (128,128)
                ),
                cv2.COLOR_BGR2RGB
            )
        ] * len(caption['captions'])
        captions += [ SYMBOL_FILTER.sub(" ",sent) for sent in caption['captions'] ]
    except:
        broken.append(caption)
        
image_data = np.array(image_data)
```

```python
from gensim.models import KeyedVectors
```

```python
w2v = KeyedVectors.load_word2vec_format("./word_embeddings/gensim_w2v.model")
```

```python
max([len(cap.split(" ")) for cap in captions])
```

```python
def vectorize(captions):
    ret = []
    for cap in captions:
        tokens = cap.lower().split(" ")
        emb = [ list(w2v[word]) for word in tokens] + ( [np.zeros(64)]*(24-len(tokens)) )
        ret.append(emb)
    return np.array(ret).astype(np.float32)
```

```python
%%time

caption_vectorized = vectorize(captions)
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras.layers import *
```

```python
### Query Encoder

def dense(x,units):
    x = Dense(units,)(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(0.3)(x)    
    return x



query_enc_inp = Input((24,64),name="query")

query_enc_l1 = LSTM(units=128,use_bias=False,return_sequences=False)(query_enc_inp)
query_enc_dr1 = Dropout(0.3)(query_enc_l1)

query_enc_d2 = dense(query_enc_dr1,128)
query_enc_d3 = dense(query_enc_d2,64)

query_enc_d4 = Dense(32,)(query_enc_d3)
query_enc_b4 = BatchNormalization()(query_enc_d4)
query_enc_out = Activation('tanh')(query_enc_b4)

image_enc_inp = Input((128,128,3),name="image")
mobilenet = keras.applications.MobileNet(weights='imagenet',input_tensor=image_enc_inp,include_top=False).output

image_enc_flat = GlobalAveragePooling2D()(mobilenet)
image_enc_flat = Flatten()(image_enc_flat)

image_enc_d3 = Dense(32,)(image_enc_flat)
image_enc_d3 = BatchNormalization()(image_enc_d3)
image_enc_out = Activation('tanh')(image_enc_d3)
        
query_encoder = keras.Model([query_enc_inp],[query_enc_out])
image_encoder = keras.Model([image_enc_inp],[image_enc_out])

model = keras.Model([query_enc_inp,image_enc_inp],  concatenate([query_enc_out , image_enc_out]))
```

```python
example = (
    caption_vectorized[0:3],
    (image_data[0:3]/255).astype(np.float32)
)
```

```python
class Loss(tf.Module):
    def __init__(self,):
        self.cosine = keras.losses.CosineSimilarity()
        self.l1 = keras.losses.MeanAbsoluteError()
        self.l2 = keras.losses.MeanSquaredError()
        
    def __call__(self,y_true,y_pred,*args,**kwargs):
        a = tf.slice(y_pred,begin=[0,0],size=[-1,32])
        b = tf.slice(y_pred,begin=[0,32],size=[-1,-1])
        return self.l2(a,b)
```

```python
image_data.shape,caption_vectorized.shape
```

```python
def train_flow(batch_size):
    while True:
        dummy = np.zeros((batch_size,64))
        index = np.arange(0,len(image_data))
        np.random.shuffle(index)

        for batch in index.reshape(-1,16):
            yield ( caption_vectorized[batch], image_data[batch] / 255 ), dummy    
```

```python
BATCH_SIZE = 16
STEPS_PER_EPOCH = len(image_data) // BATCH_SIZE
EPOCHS = 3

gen = train_flow(BATCH_SIZE)
```

```python
loss = Loss()
opt = keras.optimizers.Adam(9e-5)

model.compile(loss=loss,optimizer=opt)
```

```python
model.fit_generator(
    gen,
    steps_per_epoch=STEPS_PER_EPOCH,
    epochs=EPOCHS
)
```

```python
### generating image embeddings

def image_flow():
    index = np.arange(0,len(image_data))
    for batch in index.reshape(-1,32):
        yield image_data[batch] / 255
        
imflow = image_flow()

image_repr = image_encoder.predict_generator(imflow,steps=len(image_data)//32,verbose=1)
```

```python
def generate_query(sent):
    tokens = sent.lower().split(" ")
    return query_encoder.predict(np.array( [ [ list(w2v[word]) for word in tokens] + ( [np.zeros(64)]*(24-len(tokens)) )] ).astype(np.float32))
```
